```python
agricultura_argentina_df.cultivo = agricultura_argentina_df.cultivo.astype ('category')
agricultura_argentina_df.provincia = agricultura_argentina_df.provincia.astype ('category')
```

```python
agricultura_argentina_df.cultivo.unique()
agricultura_argentina_df.cultivo.value_counts()
```

```python
lista_cultivos_principales = agricultura_argentina_df.cultivo.value_counts().index.tolist()
lista_cultivos_principales = lista_cultivos_principales[:10]
lista_cultivos_principales
```

```python
agricultura_argentina_df.provincia.unique()
agricultura_argentina_df.provincia.value_counts()
```

```python
#De la categoría cultivo, sólo nos quedaremos con los diez más representativos y crearemos una copia del dataset con este filtro
```

```python
df_filtered=agricultura_argentina_df.loc[agricultura_argentina_df['cultivo'].isin(lista_cultivos_principales)].copy()
df_filtered.head()
```

```python
#graficamos la distribución de 4 de los proncipales cultivos por superficie cosechada
```

```python
arg_cultivos_principales=[]
mylabels=[]

for cultivo in lista_cultivos_principales:
    arg_cultivos_principales.append(df_filtered[df_filtered.cultivo==cultivo].sup_cosechada)
    mylabels.append(cultivo)

hist1= plt.hist(arg_cultivos_principales, stacked=True, label=mylabels)
plt.legend();
```

```python
fig = plt.figure()
ax = plt.axes()

plt.hist(df_filtered[df_filtered.cultivo == 'Maíz'].sup_cosechada, label='Maíz')
plt.hist(df_filtered[df_filtered.cultivo == 'Trigo total'].sup_cosechada, label='Trigo total')
plt.hist(df_filtered[df_filtered.cultivo == 'Sorgo'].sup_cosechada, label='Sorgo')
ax.set(xlabel='Superficie Cosechada', ylabel='Ocurrencia Cultivo',
       title='Estimaciones Agrícolas - Argentina')
plt.legend();
```
